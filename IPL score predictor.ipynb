```python
import pandas as pd
import numpy as np
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
match.head(5)
```

```python
delivery.head(5)
```

```python
#taking out total score per inning of every match:
total_score_df = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df.shape
```

```python
#keeping only 1st inning scores:
total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
#merging totals_score_df with match:
match_df = match.merge(total_score_df[['match_id', 'total_runs']],left_on='id', right_on='match_id')
```

```python
match_df.head()
```

```python
match_df['team1'].unique()
```

```python
#renaming team name with current names:
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
#keeping currently playing teams:
teams = [
    'Sunrisers Hyderabad', 
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings', 
    'Rajasthan Royals',
    'Delhi Capitals'
    ]
```

```python
#taking out currently playing teams:
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
#removing dl_applied matches:
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df
```

```python
#keeping required columns:
match_df = match_df[['match_id','city','total_runs']]
```

```python
match_df
```

```python
#merging match_df with delivery:
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df
```

```python
#keeping data of 1st inning:
delivery_df = delivery_df[delivery_df['inning'] == 1]
```

```python
#calculating current score:
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df
```

```python
126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
#calculating balls left after every ball:
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
(delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
#calculating crr = current runs/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df
```

```python
#calculating wickets left after every ball:
#filling na with 0 in player_dismissed:
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna(0)
```

```python
# changing dismissed player's name to 1:
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == 0 else "1")
```

```python
#converting to int:
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype(int)
```

```python
#calculating wickets left:
delivery_df['wickets'] = 10 - (delivery_df.groupby('match_id').cumsum()['player_dismissed'].values)
```

```python
#Getting runs per last 3 overs:
groups = delivery_df.groupby('match_id')

match_ids = delivery_df['match_id'].unique()
last_three = []
for id in match_ids:
    last_three.extend(groups.get_group(id).rolling(window=18).sum()['total_runs_y'].values.tolist())
```

```python
delivery_df['last_three'] = last_three
```

```python
delivery_df
```

```python
#keeping required columns:
final_df = delivery_df[['batting_team','bowling_team','city','current_score','balls_left','wickets','crr','last_three','total_runs_x']]
```

```python
final_df
```

```python
#finding out missing values:
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.corr()
```

```python
#shuffling the data:
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.head()
```

```python
#building up the model to predict scores:
from sklearn.model_selection import train_test_split
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
#importing required models:
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

#importing algorithms:
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge,Lasso
from sklearn.svm import SVR 
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

#importing metrics:
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
step1 = ColumnTransformer(transformers=[
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
step2 = StandardScaler()
```

```python
pipe = Pipeline([
    ('step1',step1),
    ('step2',StandardScaler()),
    ('step3',LinearRegression())
])

pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('R2Score: ',r2_score(y_test,y_pred))
print('MAE: ',mean_absolute_error(y_test,y_pred))
```

```python
pipe = Pipeline([
    ('step1',step1),
    ('step2',StandardScaler()),
    ('step3',Ridge())
])

pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('R2Score: ',r2_score(y_test,y_pred))
print('MAE: ',mean_absolute_error(y_test,y_pred))
```

```python
pipe = Pipeline([
    ('step1',step1),
    ('step2',StandardScaler()),
    ('step3',Lasso())
])

pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('R2Score: ',r2_score(y_test,y_pred))
print('MAE: ',mean_absolute_error(y_test,y_pred))
```

```python
pipe1 = Pipeline([
    ('step1',step1),
    ('step2',StandardScaler()),
    ('step3',RandomForestRegressor())
])

pipe1.fit(X_train,y_train)
y_pred = pipe1.predict(X_test)
print('R2Score: ',r2_score(y_test,y_pred))
print('MAE: ',mean_absolute_error(y_test,y_pred))
```

```python
pipe2 = Pipeline([
    ('step1',step1),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=500,learning_rate=0.2,max_depth=10,random_state=1))
])

pipe2.fit(X_train,y_train)
y_pred = pipe2.predict(X_test)
print('R2Score: ',r2_score(y_test,y_pred))
print('MAE: ',mean_absolute_error(y_test,y_pred))
```

```python
#exporting XGB model:
import pickle
pickle.dump(pipe2,open('pipe2.pkl','wb'))
```
